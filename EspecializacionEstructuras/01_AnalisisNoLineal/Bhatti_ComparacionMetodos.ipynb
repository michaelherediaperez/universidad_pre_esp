A continuación se presenta la comparación de soluciones a partir de diferentes métodos iterativos Newton-Raphson para la siguiente ecuación diferencial:

$$\frac{d}{dx}\left(u^2\frac{du}{dx}\right) + 4 = 0$$
$$0<x<1$$

Con las siguientes condiciones de frontera:

$$u(0) = 1;$$
$$\left(u^2\frac{du}{dx}\right)_{x=1} = 2;$$

Sabiendo que la solución exaccta es:

$$u(x) = \sqrt[3]{-6x^2+18x+1}$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Solución exacta.
xx = np.linspace(0, 1, 100)
ux = lambda x : (-6*x**2 + 18*x + 1)**(1/3)

xnod = [0, 1/3, 2/3, 1]

ux1 = [1., 1.85017, 2.17811, 2.35134]   # Bhatti 7.1.
ux2 = [1., 1.8502,  2.178,   2.34743]   # Bhatti 7.2.
ux3 = [1., 1.85222, 2.18194, 2.35602]   # Bhatti 7.3.

fig = plt.figure()
fig.set_size_inches(9,11)
ax = fig.add_subplot(111)

ax.plot(xx, ux(xx), "k-", label="Exacta")
ax.plot(xnod, ux1, "r-", label="7.1")
ax.plot(xnod, ux2, "b-", label="7.2")
ax.plot(xnod, ux3, "g-", label="7.3")

ax.set_xlabel("x")
ax.set_ylabel("u(x)")

ax.legend()
plt.show()
```
